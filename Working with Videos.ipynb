# Working with Video Captures

```python
# video path in the OS
video_path = os.path.join('data', 'video', 'tone_video.mp4')

# establisting the video capture with cap
cap = cv2.VideoCapture(video_path)

#loop through all the frames
for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    
    #read the image
    ret, frame = cap.read()
    
    #gray color transform
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    #show the image
    cv2.imshow(' ', frame)
    
    #breaking out of the loop
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
#release the cap
cap.release()
#destroy all the window
cv2.destroyAllWindows()
```

# Writing out videos

```python
cap = cv2.VideoCapture(os.path.join('data','video','tone_video.mp4'))

# Properties
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
fps = cap.get(cv2.CAP_PROP_FPS)
# Video Writer 
video_writer = cv2.VideoWriter(os.path.join('data','output','output.avi'), cv2.VideoWriter_fourcc('P','I','M','1'), fps, (width, height), isColor=False) 
                                                                                        
# Loop through each frame
for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    
    # Read frame 
    ret, frame = cap.read()
    
    # Gray transform
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Show image
    cv2.imshow('Video Player', gray)
    
    # Write out frame 
    video_writer.write(gray)
    
    # Breaking out of the loop
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

# Close down everything
cap.release()
cv2.destroyAllWindows()
# Release video writer
video_writer.release()
```
