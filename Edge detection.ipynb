# Import dependencies

```python
import cv2
import os
```

# Load Images

```python
input_path = os.path.join('data', 'Images','boy.jpg')
```

```python
img = cv2.imread(input_path)
```

```python
img
```

```python
img.shape
```

```python
#view image with the heip of matplotlib
import matplotlib.pyplot as plt
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#resizing the image t fit the cv2 window shown by image
resize_img = cv2.resize(img, (int(img.shape[1]/6), int(img.shape[0]/6)))

# view image using CV2 and destroyed by any key
cv2.imshow('frame_view', resize_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Edge Detection

```python
blurr = cv2.GaussianBlur(img, (5, 5), 0)
canny = cv2.Canny(blurr, threshold1 = 180, threshold2= 200)
```

```python
#resizing the image t fit the cv2 window shown by image
resize_img = cv2.resize(canny, (int(img.shape[1]/6), int(img.shape[0]/8)))

# view image using CV2 and destroyed by any key
cv2.imshow('frame_view', resize_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Export the image

```python
cv2.imwrite('edges.jpg', canny)
```
