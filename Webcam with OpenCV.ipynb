```python
#a function to take photo and write in the folder
def take_photo():
    cap = cv2.VideoCapture(0)
    ret, frame = cap.read()
    
    cv2.imwrite('webcam_photo.jpeg', frame)
    cap.release()
```

```python
take_photo()
```

# Rendering in the Real Time

```python
cap = cv2.VideoCapture(0)

while cap.isOpened():
    #reading the image
    ret, frame = cap.read()
    #showing te image
    cv2.imshow('webcam', frame)
    
    #closing the cam with the help of 'q' key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

#relesing frame into wild
cap.release()
#destroying all the window
cv2.destroyAllWindows()
```
